```python
import pandas as pd
import matplotlib.pyplot as plt

# 确保将此路径更改为你上传文件的正确路径
file_path = 'latent_optimization_results_with最長_svm.csv'  # 请更改为实际文件路径
results_df = pd.read_csv(file_path)

# 提取 Latent Dim 和 SVM Accuracy
latent_dims = results_df['Latent Dimension']
svm_accuracies = results_df['SVM Accuracy']  # 确保这里列名正确

# 绘制 SVM Accuracy 随 Latent Dimension 的变化
plt.figure(figsize=(10, 6))
plt.plot(latent_dims, svm_accuracies, marker='o', linestyle='-', color='b')
plt.title('SVM Accuracy vs Latent Dimension')
plt.xlabel('Latent Dimension')
plt.ylabel('SVM Accuracy')
plt.xticks(latent_dims)  # 设置 x 轴刻度
plt.grid(True)
plt.show()
```
